```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

#replaceing blank with 0 as tenure i 0 and total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", "0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df["customerID"]
```

#converted 0 and 1 value of senior cetizen to Yes/No to make easier to understand

```python
def conv(value):
    if value == 1:
        return  "Yes"
    else:
        return "No"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head()
```

```python
ax = sns.countplot(x = 'Churn' , data =df)

ax.bar_label(ax.containers[0])
plt.title("Count of Churn By Customer")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb['Churn'], labels=gb.index, autopct="%1.2f%%")
plt.title("Percentage of Churned Customer", fontsize = 10)
plt.show()
```

#from the given pie chart we can conclude that 26.54% of our customer have Churned out 
#now let's explore the reason behind it

```python
plt.figure(figsize=(4,4))
sns.countplot(x = "gender", data=df, hue="Churn")
plt.title("Churn By Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="SeniorCitizen", data=df, hue="SeniorCitizen")
plt.title("Senior Citizen by Data")
plt.show()
```

```python
ct = pd.crosstab(df['gender'], df['Churn'])

ct.plot(kind='bar', stacked=True)

plt.title("Stacked Bar Chart")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]

plt.stackplot(x, y1, y2, labels=["A", "B"])

plt.legend()
plt.title("Stack Plot")
plt.show()
```

```python
sns.countplot(x="gender", hue="Churn", data=df)

plt.title("Gender vs Churn (Stacked View)")
plt.show()
```

```python
ct = pd.crosstab(df['Contract'], df['Churn'])
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

ct_pct.plot(kind='bar', stacked=True, colormap='coolwarm')

plt.title("Contract vs Churn (%)")
plt.xlabel("Contract Type")
plt.ylabel("Percentage")
plt.xticks(rotation=0)
plt.legend(title="Churn")

plt.show()
```

```python
ct = pd.crosstab(df['PaymentMethod'], df['Churn'])
ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100

ct_pct.plot(kind='bar', stacked=True, colormap='Set2')

plt.title("Payment Method vs Churn")
plt.xticks(rotation=30)
plt.show()
```

```python
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)

plt.title("Monthly Charges vs Churn")
plt.show()
```
